```python
from operator import itemgetter
```

```python
class Node:
    def __init__(self, cost, depth, seq):
        self.cost = cost
        self.depth = depth
        self.seq = seq
        self.next = []

def getSeq(n):
    return n.seq

def generateTree(d):
    wordBank = ["cat","mat","on","sat","the"]
    wordWeight = [2,2,1,2,2]
    
    t = Node(0,0,"0")
    for i in range(len(wordBank)):
        word1 = wordBank[i]
        cost1 = wordWeight[i]
        t.next.append(Node(cost1,1,word1))
        if d > 1:
            for j in range(len(wordBank)):
                word2 = wordBank[j]
                cost2 = wordWeight[j]
                t.next[i].next.append(Node(cost1+cost2,2,word1+" "+word2))
                if d > 2:
                    for k in range(len(wordBank)):
                        word3 = wordBank[k]
                        cost3 = wordWeight[k]
                        t.next[i].next[j].next.append(Node(cost1+cost2+cost3,3,word1+" "+word2+" "+word3))
                        if d > 3:
                            for a in range(len(wordBank)):
                                word4 = wordBank[a]
                                cost4 = wordWeight[a]
                                t.next[i].next[j].next[k].next.append(Node(cost1+cost2+cost3+cost4,4,word1+" "+word2+" "+word3+" "+word4))
                                if d > 4:
                                    for b in range(len(wordBank)):
                                        word5 = wordBank[b]
                                        cost5 = wordWeight[b]
                                        t.next[i].next[j].next[k].next[a].next.append(Node(cost1+cost2+cost3+cost4+cost5,5,word1+" "+word2+" "+word3+" "+word4+" "+word5))
                                        if d > 5:
                                            for c in range(len(wordBank)):
                                                word6 = wordBank[c]
                                                cost6 = wordWeight[c]
                                                t.next[i].next[j].next[k].next[a].next[b].next.append(Node(cost1+cost2+cost3+cost4+cost5+cost6,5,word1+" "+word2+" "+word3+" "+word4+" "+word5+" "+word6))
    return t
```

```python
def bfs(t):
    q = []
    q.append(t)
    while q:
        n = q.pop(0)
        for nxt in n.next:
            q.append(nxt)
        print(n.seq)
        if n.seq == "the cat sat" or n.seq == "the cat on the mat sat" or n.seq == "the cat sat on the mat":
            return
```

```python
def uni(t,c):
    nodes = [(t.seq,t.cost)]
    q = []
    q.append(t)
    isRoot = True
    while q:
        n = q.pop(0)
        for nxt in n.next:
            q.append(nxt)
        if isRoot == False:
            nodes.append((n.seq,n.cost))
        isRoot = False
    nodes = sorted(nodes, key=itemgetter(0))
    for ci in range(c):
        for (seq,cost) in nodes:
            if cost == ci:
                print(seq)
                if seq == "the cat sat" or seq == "the cat on the mat sat" or seq == "the cat sat on the mat":
                    return
```

```python
def ite(t,d):
    nodes = [(t.seq,t.depth)]
    q = []
    q.append(t)
    isRoot = True
    while q:
        n = q.pop(0)
        for nxt in n.next:
            q.append(nxt)
        if isRoot == False:
            nodes.append((n.seq,n.depth))
        isRoot = False
    nodes = sorted(nodes, key=itemgetter(0))
    for di in range(d):
        for (seq,depth) in nodes:
            if depth == di:
                print(seq)
                if seq == "the cat sat" or seq == "the cat on the mat sat" or seq == "the cat sat on the mat":
                    return
```
